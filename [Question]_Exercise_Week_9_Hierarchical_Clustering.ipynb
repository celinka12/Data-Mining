```python
# Encoding Gender
le = LabelEncoder()
df_imputer['Arah Hadap Rumah'] = le.fit_transform(df_imputer['Arah Hadap Rumah'])
df_imputer['Posisi Rumah'] = le.fit_transform(df_imputer['Posisi Rumah'])

df_imputer['Terjual/Belum'] = df_imputer['Terjual/Belum'].map({'Terjual': 0, 'Belum': 1})
df_imputer['Area Category'] = df_imputer['Area Category'].map({'Sangat Premium': 1, 'Premium': 2, 'Standard': 3, 'Below Standard': 4})
df_imputer['Lebar Jalan Depan Rumah (ROW)'] = df_imputer['Lebar Jalan Depan Rumah (ROW)'].map({'< 1 Mobil': 1, '1-2 Mobil': 2, '> 2 Mobil': 3})
df_imputer.head()
```

To replace incorrect or invalid data in the column "Harga Penawaran (dari Owner)", do the following:

Replace empty values (" ") with "10500000000".

Replace values with commas like "10.500.000.000,00" with "10500000000".

```python
# Step 1: Replace empty values (" ") with "10500000000"
df_imputer['Harga Penawaran (dari Owner)'] = df_imputer['Harga Penawaran (dari Owner)'].replace(' ', '10500000000')

# Step 2: Remove commas and periods, and replace with the correct format
df_imputer['Harga Penawaran (dari Owner)'] = df_imputer['Harga Penawaran (dari Owner)'].str.replace('.', '').str.replace(',', '').replace('', '10500000000')

# Convert the column to integer type if necessary
df_imputer['Harga Penawaran (dari Owner)'] = df_imputer['Harga Penawaran (dari Owner)']
```

## K-Means Clustering

Using the elbow method, determine the number of clusters to be formed.

Use the elbow range range(1, 11)

```python
# set the parameter for evaluating K
sse=[]
# check the inertia for each K
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='random', max_iter=300, random_state=42)
    kmeans.fit(df_imputer)
    sse.append(kmeans.inertia_)

# manually visualize the elbow method using plotlib
plt.plot(range(1, 11), sse, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('SSE')
plt.title("Elbow Index")
plt.show()
```

```python
# choosing best K
from kneed import KneeLocator

findknee = KneeLocator(range(1,11), sse, curve="convex", direction="decreasing")
best_k = findknee.knee
print("the best k for your model is ", best_k)
```

Train the K-Means model on the dataset with the number of clusters you have determined.

```python
kmeans = KMeans(n_clusters=best_k, init='random', max_iter=300, random_state=42)
kmeans.fit(df_imputer)
labels = kmeans.labels_
```

```python
df_imputer['KMeans'] = kmeans.labels_

# Now inverse transform the scaled numeric columns to retrieve the original values
df_imputer[numeric_cols] = scaler.inverse_transform(df_imputer[numeric_cols])

# Display the first few rows of the DataFrame to veriacfy the original values for numeric columns
df_imputer.head()
```

Display the cluster summary in the form of a dataframe:

1. For categorical columns: mode (use pd.Series.mode to find the mode; you do not need to use " " for this attribute).
2. For continuous columns: mean.
3. Cluster results: count.

```python
# aggregation
def mode(series):
    return series.mode()[0] if not series.mode().empty else np.nan
agg_dict = {
    'Area Category': [mode, 'min', 'max'],
    'Luas Tanah (m2)': ['mean', 'min', 'max'],
    'Luas Bangunan (m2)': ['mean', 'min', 'max'],
    'Jumlah Kamar': ['mean', 'min', 'max', mode],
    'Jumlah Kamar Mandi': ['mean', 'min', 'max', mode],
    'Tingkat/Lantai ': ['mean', 'min', 'max', mode],
    'Harga Penawaran (dari Owner)': ['mean', 'min', 'max'],
    'Terjual/Belum': mode,
    'Arah Hadap Rumah': mode,
    'Posisi Rumah': mode,
    'Lebar Jalan Depan Rumah (ROW)': mode
}
agg_result = df_imputer.groupby('KMeans').agg(agg_dict)
agg_result.columns = [f"{col[0]} ({col[1]})" if col[1] else col[0] for col in agg_result.columns]
agg_result_transposed = agg_result.transpose()
agg_result_styled = agg_result_transposed.style.format(precision=2)
agg_result_styled
```

## Hierarchical Clustering
Hierarchical clustering uses the agglomerative method.

```python
# The image below is not the final result, if yours looks different, that’s okay.
from sklearn.metrics import silhouette_score

silhouette_scores = []

for i in range(2, 11):
    cluster_labels = fcluster(linkage(df_imputer, method='ward'), i, criterion='maxclust')
    silhouette_avg = silhouette_score(df_imputer, cluster_labels)
    silhouette_scores.append(silhouette_avg)

# Plot Silhouette Score = the higher the better
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score")
plt.show()
```

Perform hierarchical clustering with the Euclidean distance metric and set the number of clusters based on silhouette score.

```python
# agglomerative clustering
silhouette_scores = []
for k in range(2, 10):  # Mencoba jumlah cluster
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(df_imputer)  # X = dataset
    score = silhouette_score(df_imputer, labels)
    silhouette_scores.append(score)
    print(f"Jumlah Cluster: {k}, Silhouette Score: {score}")
```

**Hierarchical Visualization**

Single Linkage

```python
# Min Method
n_clusters = 3
# Change 'max' to 'single' for single linkage
linked = linkage(df_imputer, method='single', metric='euclidean')
cluster_labels = fcluster(linked, t=n_clusters, criterion='distance')

color_threshold = linked[-(n_clusters-1), 2]

# Plot dendrogram dengan color_threshold
plt.figure(figsize=(10, 7))
sch.dendrogram(linked,
           color_threshold=color_threshold,
           above_threshold_color='orange',
           labels=np.arange(1, df_imputer.shape[0] + 1))
plt.title("Dendrogram with Min Method")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

complete linkage

```python
# Max Method
n_clusters = 3
linked = linkage(df_imputer, method='complete', metric='euclidean')
cluster_labels = fcluster(linked, t=n_clusters, criterion='distance')

color_threshold = linked[-(n_clusters-1), 2]
# Plot dendrogram dengan color_threshold
plt.figure(figsize=(10, 7))
sch.dendrogram(linked,
           color_threshold=color_threshold,
           above_threshold_color='green',
           labels=np.arange(1, df_imputer.shape[0] + 1))
plt.title("Dendrogram with Complete Method")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```

average linkage

```python
# Avg Method
n_clusters = 3
linked = linkage(df_imputer, method='average', metric='euclidean')
cluster_labels = fcluster(linked, t=n_clusters, criterion='distance')

color_threshold = linked[-(n_clusters-1), 2]
# Plot dendrogram dengan color_threshold
plt.figure(figsize=(10, 7))
sch.dendrogram(linked,
           color_threshold=color_threshold,
           above_threshold_color='orange',
           labels=np.arange(1, df_imputer.shape[0] + 1))
plt.title("Dendrogram with AVERAGE Method")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```
